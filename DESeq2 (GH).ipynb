# Create a metadata DF

```r
metadata <- data.frame(
  condition = factor(c("LPS_IFNr_Wy14643", "LPS_IFNr", "PBS", "LPS_IFNr_Wy14643", "LPS_IFNr", "PBS")),
  batch = factor(c("SRR17869115",	"SRR17869118",	"SRR17869121", "SRR17869116",	"SRR17869119",	"SRR17869122")),
  row.names = colnames(counts)
)
```

# Part I - Create a DESeq2 data object

```r
dds <- DESeqDataSetFromMatrix(
  countData = counts,
  colData = metadata,
  design = ~ condition
)
```

# Part II - Run DESeq2 Analysis

```r
dds <- DESeq(dds)
```

## Extract results

```r
dds
```

```r
# Compare treated conditions
res <- results(dds, contrast = c("condition", "LPS_IFNr_Wy14643", "LPS_IFNr"))
```

```r
# Add gene annotations to the 'res' object
res$gene_symbol <- mcols(dds)[rownames(res), "gene_symbol"]
res$gene_description <- mcols(dds)[rownames(res), "gene_description"]
```

```r
res
```

```r
summary(res)
```

## Filter and explore results

```r
resOrdered <- res[order(res$padj),]  # Order by adjusted p-value
summary(res)  # Get a summary of the results
resSig <- subset(resOrdered, padj < 0.05)  # Filter for significant genes
```

## Plot counts

```r
# Plot expression levels of  most significantly differentially expressed gene across different experimental conditions
plotCounts(dds, gene=which.min(res$padj), intgroup="condition")
```

## Plot results

```r
# MA plot
plotMA(res, ylim=c(-2,2))
```

```r
# Using gene_symbol for labels in EnhancedVolcano
EnhancedVolcano(res,
    lab = res$gene_symbol, # Use the newly added gene_symbol column
    x = 'log2FoldChange',
    y = 'pvalue')
```

## Heatmap

```r
# Get normalized counts for significant genes
normalized_counts <- counts(dds, normalized=TRUE)[rownames(resSig), ]

# Subset for the samples you want to include in the heatmap (e.g., all samples)
heatmap_data <- normalized_counts[, colnames(dds)]
```

```r
pheatmap(heatmap_data,
         cluster_rows = TRUE,
         cluster_cols = TRUE,
         scale = "row", # Scale by row to visualize expression patterns
         annotation_col = metadata, # Add metadata annotation to columns
         show_rownames = FALSE) # Hide row names for better visualization
```

# Part III - GO enrichment analysis

```r
# Prepare Input
significant_genes <- rownames(resSig) # Assuming rownames are Entrez IDs

# If rownames are gene symbols, convert to Entrez IDs:
# significant_genes <- mapIds(org.Mm.eg.db, keys=rownames(resSig),
#                             column="ENTREZID", keytype="SYMBOL")
```

```r
head(significant_genes)
```

```r
# Run GO enrichment
ego <- enrichGO(gene          = significant_genes,
                OrgDb         = org.Mm.eg.db, # Replace with appropriate organism database
                keyType       = 'ENTREZID',
                ont           = "ALL", # Options: BP (Biological Process), CC (Cellular Component), MF (Molecular Function)
                pAdjustMethod = "BH",
                pvalueCutoff  = 0.05,
                qvalueCutoff  = 0.05,
                readable      = TRUE) # Add readable=TRUE
```

```r
# View results
head(ego,3)
```

```r
ontology_counts <- table(ego$ONTOLOGY)
print(ontology_counts)
```

## Dotplot

```r
# Subset results according to ontology categories
ego_BP <- ego
ego_BP@result <- subset(as.data.frame(ego), ONTOLOGY == "BP")

ego_CC <- ego
ego_CC@result <- subset(as.data.frame(ego), ONTOLOGY == "CC")

ego_MF <- ego
ego_MF@result <- subset(as.data.frame(ego), ONTOLOGY == "MF")

# Now you can use ego_BP for dotplot, keeping the enrichGO class
plot_BP <- dotplot(ego_BP, showCategory = 20)
plot_CC <- dotplot(ego_CC, showCategory = 20)
plot_MF <- dotplot(ego_MF, showCategory = 20)

dotplot <- cowplot::plot_grid(plot_BP, plot_CC, plot_MF, ncol=3, labels=c("BP", "CC", "MF"))
```

```r
# Save the combined plot with specified dimensions
ggsave(filename = "combined_dotplot.png", dotplot, width = 36, height = 12)
```
