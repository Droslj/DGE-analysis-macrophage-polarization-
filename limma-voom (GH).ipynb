# Create design matrix & contrasts
The design matrix specifies the experimental design, defining which samples belong to which groups.

```r
group <- dge$samples$group
```

```r
# Create a design matrix with valid column names
design <- model.matrix(~0+group)
colnames(design) <- make.names(colnames(design)) # Use make.names to create valid R names
design
```

```r
# Create contrasts with valid level names
contr.matrix <- makeContrasts(
   C1 = groupLPS_IFNr_Wy14643 - groupLPS_IFNr,
   C2 = groupLPS_IFNr_Wy14643 - groupPBS,
   C3 = groupLPS_IFNr - groupPBS,
   levels = colnames(design)) # Use valid column names from the design matrix
contr.matrix
```

# Perform additional filtering

```r
# Filter DGEList object
keep <- filterByExpr(dge, group=group)  # 'group' is your experimental design factor
dge <- dge[keep,,keep.lib.sizes=FALSE]
```

# Perform voom transformation

```r
# Voom transformation
v <- voom(dge, design = model.matrix(~group, data = dge$samples), plot = TRUE)
```

```r
head(v)
```

```r
# Plot distribution after Voom transformation
boxplot(v$E)
```

# Fit linear model

```r
fit <- lmFit(v, design)
```

```r
summary(fit)
```

# Apply contrasts

```r
fit2 <- contrasts.fit(fit, contr.matrix)
```

# Perform differential expression analysis

This line applies the eBayes function from the limma package to the linear model fit object fit2. The eBayes function computes moderated t-statistics, moderated F-statistics, and log-odds of differential expression by shrinking the estimated variances towards a common value. This step is crucial for improving the stability and power of the statistical tests, especially with a limited number of replicates.

```r
fit2 <- eBayes(fit2)
```

```r
results <- topTable(fit2, coef = 1, adjust.method = "BH", number = Inf)  # coef = 1 for the first contrast
```

```r
summary(fit2)
```

```r
results
```

```r
significant_degs <- results[results$adj.P.Val < 0.05 & abs(results$logFC) > 2, ]
```

```r
# Print results
head(significant_degs)
```

```r
dim(significant_degs)
```

```r
# Extract expression data for top genes
expression_data <- v$E[top_genes, ]
```

```r
expression_data
```

```r
# Test if all the genes have corresponding symbols
missing_genes <- rownames(expression_data)[!rownames(expression_data) %in% dge$genes$ENTREZID]
if(length(missing_genes) > 0) {
    print("Warning: The following genes from top_genes were not found in dge$genes:")
    print(missing_genes)
} else {
    print("All top genes found in dge$genes.")
}
```

```r
# or
present_genes <- intersect(rownames(expression_data), dge$genes$ENTREZID)
present_genes
```

```r
# Get the original gene identifiers from the expression data
original_gene_ids <- rownames(expression_data)

# Now, find the corresponding rows in the dge$genes data frame based on the ENTREZID column
# We need to make sure that the original gene IDs from expression_data are actually present
# in the ENTREZID column of dge$genes.
# We can create a named vector mapping ENTREZID from dge$genes to SYMBOL
entrez_to_symbol_map <- dge$genes$SYMBOL
names(entrez_to_symbol_map) <- dge$genes$ENTREZID

# Use the original gene IDs to look up symbols in the map
# If an original_gene_id is not in the map (i.e., not a mapped ENTREZID), it will result in NA
symbols_for_plotting <- entrez_to_symbol_map[original_gene_ids]

# Create a data frame for plotting
plot_data <- data.frame(
    Gene = rep(symbols_for_plotting, ncol(expression_data)), # Use the looked-up gene symbols here
    Expression = as.vector(expression_data),
    Group = rep(dge$samples$group, each = nrow(expression_data))
)

plot_data
```

```r
# Create stripchart using ggplot2
ggplot(plot_data, aes(x = Group, y = Expression)) +
    geom_jitter(width = 0.2, alpha = 0.9, shape = 16, size = 2) +
    facet_wrap(~ Gene, scales = "free_y", ncol = 2) +  # Separate plots for each gene using Gene (symbols)
    labs(title = "Stripcharts of Top DEGs", x = "Group", y = "Expression") +
    theme(axis.text.x = element_text(angle = -30, vjust = 1, hjust = 0)) +
    xlab(NULL) +
    ylab(NULL) +
    theme_bw()  # Optional: Use a black and white theme
```
